```python
import io
import pandas as pd
import seaborn as sns
import pandas_profiling as pp
import matplotlib.pyplot as plt
import random
```

```python
dfProducts = pd.read_table("D:/SEBLU/2ndYear/BigDataManagement/Homework-3/bi_product.txt", index_col = 0, delimiter = ";", decimal = ".")
```

```python
dfProducts
```

```python
dfSalesFact = pd.read_table("D:/SEBLU/2ndYear/BigDataManagement/Homework-3/bi_salesFact.txt", index_col = 0, delimiter = ";", decimal = ".")
```

```python
dfSalesFact
```

```python
merged_tables = pd.merge(dfProducts, dfSalesFact, on = "ProductID", how = "inner")
```

```python
merged_tables
```

```python
prof_report = pp.ProfileReport(merged_tables)
```

```python
prof_report.to_file(output_file="profile_report_.html")
```

```python
def function_to_increase(df):
    df["RevenueIncreased"] = df.apply(lambda row : row["Revenue"] * (random.randint(5, 15) / 100 + 1), axis = 1)
    return(df)
```

```python
increased_table = function_to_increase(merged_tables)
```

```python
increased_table[:5]
```

```python
tables_grouped = increased_table.groupby(["Category"])
tables_grouped.sum()
```

```python
print("Mean result:")
print(tables_grouped["Revenue"].mean())
print("Median result:")
print(tables_grouped["Revenue"].median())
```

```python
a = increased_table["Revenue"].skew()
if increased_table["Revenue"].skew() > 0 :
    print("Skewness has a positive value: ({val}), and that means: mode < median < mean, and distribution has thicker right tail!".format(val=a))
elif increased_table["Revenue"].skew() < 0 :
    print("Skewness has a negative value ({val}), and that means: mode > median > mean, and distribution has thicker left tail!".format(val=a))
else:
    print("Distribution is even!")
```
